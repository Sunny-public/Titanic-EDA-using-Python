# Exploratory Data Analysis (EDA)
Titanic Dataset

This notebook performs data exploration, visualization, and insight extraction.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

train.head()
```

## Dataset Overview

```python
train.info()
```

## Statistical Summary

```python
train.describe()
```

## Missing Values

```python
train.isnull().sum()
```

## Categorical Analysis

```python
print(train['Sex'].value_counts())
print(train['Pclass'].value_counts())
print(train['Embarked'].value_counts())
```

## Histograms

```python
train.hist(figsize=(10,8))
plt.show()
```

## Boxplot (Fare vs Survival)

```python
sns.boxplot(x='Survived', y='Fare', data=train)
plt.show()
```

## Scatterplot (Age vs Fare)

```python
sns.scatterplot(x='Age', y='Fare', data=train)
plt.show()
```

## Correlation Heatmap

```python
corr = train.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```

## Pairplot

```python
sns.pairplot(train.select_dtypes(include=['int64','float64']))
plt.show()
```

## Key Insights

- Fare is highly skewed.
- Higher class passengers have higher survival rate.
- Strong relationship between Fare, Pclass, and Survival.
